```python
import cv2

image = cv2.imread("AstonMcar.jpg")
print("Image type",type(image))
cv2.imshow("Image name", image)

cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(image)
plt.show()
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print("Image shape", image_gray.shape)
plt.imshow(image_gray, cmap="gray")
plt.show()
```

```python
small_image = cv2.resize(image_rgb, (100, 100))
plt.imshow(small_image)
plt.show()
```

```python
# ksize
ksize = (15, 15)

#Using cv2.blur() method
image_blured = cv2.blur(image_rgb, ksize)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (15, 10))
ax1.imshow(image_rgb)

ax2.imshow(image_blured)

plt.show()
```

```python
# Canny edge detector, has to do with detecting edges
image_edge = cv2.Canny(image_rgb, 125, 175) # two thresholds
plt.imshow(image_edge)

plt.show()
```

```python
import cv2

# represents the top left corner of rectangle
start_point = (99, 90) # right, down

# reprents the bottom right corner of rectangle
end_point = (1620, 1010)

# Red color in RGB
color = (255, 0, 0)

# Line thickness of 2 px
thickness = 5

# Using cv2.rectangle() method
image_rectangle = cv2.rectangle(image_rgb, start_point, end_point, color, thickness)
plt.imshow(image_rectangle)
plt.show()
```
